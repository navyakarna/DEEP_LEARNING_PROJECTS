```python
import numpy as np
from skimage.io import imread, imshow
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = imread('/content/2350.jpg')
print('original image')
imshow(image)
```

```python
from skimage.transform import rotate
rotated = rotate(image, angle=30)
print('Rotated image')
imshow(rotated)
```

```python
from skimage.transform import warp, AffineTransform
transform = AffineTransform(translation=(40,40))
Shift = warp(image, transform)

plt.imshow(Shift)
print('Normal Shift')
```

```python
warpshift = warp(image, transform, mode= 'edge')
plt.imshow(warpshift)
plt.title('Warp Shift')
```

```python
fliplr = np.fliplr(image)
plt.imshow(fliplr)
```

```python
flipud = np.flipud(image)
plt.imshow(flipud)
```

```python
from skimage.util import random_noise
noisyRandom = random_noise(image, var=0.2)
plt.imshow(noisyRandom)
```

```python
noisyGaussian = random_noise(image, mode='gaussian')
plt.imshow(noisyGaussian)
```

```python
from skimage.filters import gaussian
blurred = gaussian(image, sigma=2.5)
plt.imshow(blurred)
```
